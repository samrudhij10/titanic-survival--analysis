```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('titanic.csv')
df.head(5)
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
# 1. Who survived more: males or females?
df.groupby('Sex')['Survived'].mean()
```

```python
# 2. Did passenger class affect survival?
df.groupby('Pclass')['Survived'].mean()
```

```python
# 3. Survival rate by age group
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0,12,19,35,60,100],
    labels=['Child','Teen','Young Adult','Adult','Senior']
)

df.groupby('Age_Group')['Survived'].mean()
```

```python
# Bar Chart: Survival by Gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
# Bar Chart: Survival by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Histogram of Passenger Ages
plt.hist(df['Age'], bins=20)

plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")

plt.show()
```

```python
1. Females survived more than males.
2. First-class passengers had higher survival rates.
3. Young adults were the largest passenger group.
4. Lower passenger class had lower survival chances.
```
